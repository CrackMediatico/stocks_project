```python
# Se impòrtan las librería necesarias
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima
from sklearn.metrics import r2_score
```

## Importación de datos:

Se importan los datos que previamente han sido ingestados y procesados, se transforma el campo correspondiente a fecha para que sean correctamente procesados y se ordenan los datos cronológicamente

```python
df_comp = pd.read_csv('processed_data/data_input.csv', header='infer', parse_dates=True, infer_datetime_format=True)
```

```python
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
```

```python
df_comp = df_comp.sort_values('date')
df_comp = df_comp.set_index('date')
```

```python
df_comp.head()
```

## Ejercicio 1:
**Estudio de la predictibilidad de la evolución de una criptomoneda: Bitcoin**

Seleción de las series de datos que se emplearán para el ejercicio:
- Valor de cierre de Bitcoin diario
- Valor de cierre de Ethereum diario
- Valor de cierre del cambio Euro - Dollar diario

```python
# df_comp1.columns
column_list1 = ['Bitcoin_close_value', 'Ethereum_close_value', 'EURO-DOLLAR_close_value']
df_comp1 = df_comp[column_list1]
df_comp1.head()
```

Se realiza el limpiado de los datos y se asegura que no existan valores nulos.

```python
df_comp1 = df_comp1.dropna()
```

```python
df_comp1.isnull().sum()
```

```python
df_comp1.info()
```

Representación de las serie elegidas. Se puede observar como los datos presentan tendencias por lo que es necesario convertirlas a series de datos estacionarias.

```python
# Plot
fig, axes = plt.subplots(nrows=3, ncols=1, dpi=120, figsize=(10,6))
for i, ax in enumerate(axes.flatten()):
    data = df_comp1[df_comp1.columns[i]]
    ax.plot(data, color='red', linewidth=1)
    # Decorations
    ax.set_title(df_comp1.columns[i])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines["top"].set_alpha(0)
    ax.tick_params(labelsize=6)

plt.tight_layout();
```

Se transforman los datos a series donde podamos analizar el cambio diario de las distintas series temporales.

```python
df_comp1['Bitcoin_close_value'] = df_comp1.Bitcoin_close_value.pct_change(1)*100
df_comp1['Ethereum_close_value'] = df_comp1.Ethereum_close_value.pct_change(1)*100
df_comp1['EURO-DOLLAR_close_value'] = df_comp1['EURO-DOLLAR_close_value'].pct_change(1)*100
```

```python
# Plot
fig, axes = plt.subplots(nrows=3, ncols=1, dpi=120, figsize=(10,6))
for i, ax in enumerate(axes.flatten()):
    data = df_comp1[df_comp1.columns[i]]
    ax.plot(data, color='red', linewidth=1)
    # Decorations
    ax.set_title(df_comp1.columns[i])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines["top"].set_alpha(0)
    ax.tick_params(labelsize=6)

plt.tight_layout();
```

**Entrenamiento del modelo:**

Se seleccionan el conjunto de los datos a excepción de los últimos 7 días para entrenar el modelo, reservando estos últimos 7 días para la comprobación de la efectividad del modelo.

Se ha empleado la librería pmdarima para entrenar el modelo. Con la función auto_arima podemos dejar a la librería que seleccione el mejor modelo autorregresivo de forma autónoma.

```python
size = int(len(df_comp1)-7)
df, df_test = df_comp1.iloc[:size], df_comp1.iloc[size:]
```

**Primer modelo:** Modelo sobre una única variable.

Modelo obtenido: ARIMA(2, 0, 2)

```python
model_auto = auto_arima(df['Bitcoin_close_value'][1:])
```

```python
model_auto
```

```python
model_auto.summary()
```

**Segundo modelo:** Inclusión de Ethereum y cambio EURO - DOLLAR como variables exógenas

Modelo obtenido: ARIMA(1, 0, 3)(3, 0, 1)

```python
model_auto2 = auto_arima(df['Bitcoin_close_value'][1:], exogenous = df[['Ethereum_close_value', 'EURO-DOLLAR_close_value']][1:], m = 5,
                       max_order = None, max_p = 7, max_q = 7, max_d = 2, max_P = 4, max_Q = 4, max_D = 2,
                       maxiter = 50, alpha = 0.05, n_jobs = -1, trend = 'ct', information_criterion = 'oob',
                       out_of_sample_size = 7)
```

```python
model_auto2.summary()
```

**Tercer modelo:** Inclusión de Ethereum como variable exógena

Modelo obtenido: ARIMA(3, 0, 2)(2, 0, 1)

```python
model_auto3 = auto_arima(df['Bitcoin_close_value'][1:], exogenous = df[['Ethereum_close_value']][1:], m = 5,
                       max_order = None, max_p = 7, max_q = 7, max_d = 2, max_P = 4, max_Q = 4, max_D = 2,
                       maxiter = 50, alpha = 0.05, n_jobs = -1, trend = 'ct', information_criterion = 'oob',
                       out_of_sample_size = 7)
```

```python
model_auto3.summary()
```

**Predicción:**

Del análisis de los resultados de los tres modelos propuestos, observamos que el tercer modelo es el que cuenta con unos valores más bajos de AIC y BIC. Sin embargo, en todos los casos se observa también que los valores P son muy elevados, por lo que no podemos contar con una confianza estatística razonable.

Debido a esto, se ha optado por realizar la predicción con el modelo entrenado únicamente con la propia variable, al considerarse que el resto de series no tienen una correlación con la variable estudiada.

```python
prediction1 = pd.DataFrame(model_auto.predict(n_periods = len(df_test)),index=df_test.index)
prediction1.columns = ['predicted_increase']
prediction1
```

En la representación gráfica de los valores de entrenamiento, test y predicción podemos ver cómo se cumple lo anteriormente expuesto, no obteniendo una predicción confiable.

```python
plt.figure(figsize=(8,5))
plt.plot(df['Bitcoin_close_value'][len(df)-60:],label="Training")
plt.plot(df_test['Bitcoin_close_value'],label="Test")
plt.plot(prediction1,label="Predicted")
plt.legend(loc = 'upper left')
plt.show()
```

```python

df_test['predicted_increase'] = prediction1
r2_score(df_test['Bitcoin_close_value'], df_test['predicted_increase'])
```
